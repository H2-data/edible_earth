###**Scraping Process**

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)',
    'Accept-Language': 'en-US,en;q=0.9',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
    'Referer': 'https://google.com',
    'DNT': '1'
}

murlist = [
    'https://museumofedible.earth/REDACTED',
    'https://museumofedible.earth/REDACTED',
    'https://museumofedible.earth/REDACTED',
    'https://museumofedible.earth/REDACTED',
    'https://museumofedible.earth/REDACTED',
    'https://museumofedible.earth/REDACTED',
    'https://museumofedible.earth/REDACTED',
    'https://museumofedible.earth/REDACTED'
]
```

```python
urlist = []

for item in murlist:
    print(f"Requesting: {item}")
    response = requests.get(item, headers = headers)
    soup = BeautifulSoup(response.text, 'html.parser')
    product_links = [a['href'] for a in soup.select('a.woocommerce-LoopProduct-link') if 'href' in a.attrs]
    print(f"Found {len(product_links)} product links on {item}")
    urlist.extend(product_links)
```

```python
#The output for the above code will not be shown for security reasons.
```

```python
all_tags = []

for url in urlist:
    response = requests.get(url, headers = headers)
    soup = BeautifulSoup(response.text, 'html.parser')

    tag_links = soup.select('a[rel="tag"]')
    tag_texts = [tag.get_text(strip = True) for tag in tag_links]

    all_tags.append(tag_texts)


item_names = []

for url in urlist:
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    item_name_tag = soup.select_one('h1.entry-title')
    if item_name_tag:
        item_name = item_name_tag.get_text(strip=True)
    else:
        item_name = "No name found"

    item_names.append(item_name)
```

```python
df = pd.DataFrame({'tags': all_tags}, index = item_names)
df = df.reset_index()
df.head()
```

```python
df.to_csv('product_urls.csv', index = False)
from IPython.display import FileLink
FileLink('product_urls.csv')
```
